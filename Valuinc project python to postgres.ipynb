```python
data_df = pd.read_csv('transaction.csv', sep=';')
```

```python
data_df.head()
```

```python
data_df['SalesPerTransac'] = (data_df['SellingPricePerItem'] * data_df['NumberOfItemsPurchased']).round(2)
data_df['CostsPerTransac'] = (data_df['CostPerItem'] * data_df['NumberOfItemsPurchased']).round(2)
data_df['ProfitPerTransac'] = (data_df['NumberOfItemsPurchased'] * (data_df['SellingPricePerItem'] - data_df['CostPerItem'])).round(2)
data_df['Markup'] = (data_df['ProfitPerTransac'] / data_df['CostsPerTransac']).round(2)
```

```python
data_df["Month"]
```

```python
data_df["Month1"] = [datetime.strptime(x, '%b').month for x in data_df["Month"]]
```

```python
data_df['Date'] = data_df['Month1'].apply(str) + '-' + data_df['Day'].apply(str)  + '-' + data_df['Year'].apply(str) + ' ' + data_df['Time']
```

```python
data_df.head()
```

```python
split_col = data_df['ClientKeywords'].str.split(',' , expand = True)
```

```python
data_df['ClientAgeStat'] = split_col[0]
data_df['Businesstype'] = split_col[1]
data_df['LengthofContract'] = split_col[2]
```

```python
# using the replace func
data_df['ClientAgeStat'] = data_df['ClientAgeStat'].str.replace('[','')
data_df['Businesstype'] = data_df['Businesstype'].str.replace("'",'')
data_df['LengthofContract'] = data_df['LengthofContract'].str.replace(']','')
```

```python
data_df['ClientAgeStat'] = data_df['ClientAgeStat'].str.replace("'",'')
data_df['LengthofContract'] = data_df['LengthofContract'].str.replace("'",'')
```

```python
data_df['ItemDescription'] = data_df['ItemDescription'].str.lower()
```

```python
data_df.head()
```

```python
season_df = pd.read_csv('value_inc_seasons.csv', sep = ';')
```

```python
season_df.head()
```

```python
x_df = pd.merge(data_df, season_df, on = 'Month')
```

```python
x_df = x_df.drop(columns = ['ClientKeywords', 'Year', 'Month', 'Day','Time', 'Month1']) 
```

```python
#export to csv
x_df.to_csv("valueinc_cleaned.csv", index = False)
```
